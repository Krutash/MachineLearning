```python
import os
import io
import numpy as np
from pandas import DataFrame
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

def readFiles(path):
    for root, dirnames, filenames, in os.walk(path):
        for filename in filenames:
            path = os.path.join(root, filename)
            
            inBody = False
            lines= []
            f = io.open(path, 'r', encoding = 'latin1')
            for line in f:
                if inBody:
                    lines.append(line)
                elif line == '\n':
                    inBody = True
            f.close()
            message = '\n'.join(lines)
            yield path, message
                
def dataFrameFromDirectory(path, classification):
    rows = []
    index = []
    for filename, message in readFiles(path):
        rows.append({'message': message, 'class': classification})
        index.append(filename)
        
    return DataFrame(rows, index = index)

data = DataFrame({'message': [], 'class': []})
data = data.append(dataFrameFromDirectory('C:\\Users\\utkro\\Desktop\\utk\\aaaaaaaaaaa data\\emails\\ham', 'ham'))
data = data.append(dataFrameFromDirectory('C:\\Users\\utkro\\Desktop\\utk\\aaaaaaaaaaa data\\emails\\spam', 'spam'))
```

```python
from sklearn.model_selection import train_test_split as tts

trainX, testX = tts(data, test_size = 0.2, shuffle = True)
print(testX)
```

```python
data.tail()
```

```python
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(trainX['message'].values)

classifier = MultinomialNB()
targets = trainX['class'].values
classifier.fit(counts, targets)

```

```python
testing_counts = vectorizer.transform(testX['message'].values)
predictions = classifier.predict(testing_counts)
predictions
```

```python
outPredict = open("outPUT_predictions.txt", 'w')
for classi in predictions :
    outPredict.write(classi)
    outPredict.write("\n")
outPredict.close()

outActual = open("outPUT_actual.txt", 'w')
for classi in testX['class'].values:
    outActual.write(classi)
    outActual.write("\n")
outActual.close()

outTrain = open("outPUT_Train.txt", 'w')
for classi in trainX['class'].values:
    outTrain.write(classi)
    outTrain.write("\n")
outTrain.close()
```

```python
file1_predict = open('outPUT_predictions.txt', 'r')
file2_actual = open('outPUT_actual.txt', 'r')
count = 0
correct = 0
while True:
    count += 1
    value_pred = file1_predict.readline()
    value_act = file2_actual.readline()
    
    if not value_pred:
        break
    if value_pred == value_act :
        correct += 1
file1_predict.close()
file2_actual.close()
percent = (correct/count)*100
print(percent)
```
